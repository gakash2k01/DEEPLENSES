# Imports

```python
# Importing required modules
import warnings
import os
import seaborn as sns
from tqdm import tqdm
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

# EDA

### Class count

```python
counts = {}
print(os.listdir('dataset/dataset/train'))
for i in os.listdir('dataset/dataset/train'):
    if i == '.DS_Store':
        continue
    counts[i] = len(os.listdir(f'dataset/dataset/train/{i}'))
```

```python
sns.barplot(x = list(counts.keys()), y = list(counts.values()))
```

The classes are perfectly balanced.

### Unique sizes

```python
sizes = []
for i in tqdm(os.listdir('dataset/dataset/train/no')):
    data = np.load(f'dataset/dataset/train/no/{i}')
    sizes.append(str(data.shape))
sizes = pd.Series(sizes)
print(sizes.value_counts())
```

```python
sizes_sphere = []
for i in tqdm(os.listdir('dataset/dataset/train/sphere')):
    data = np.load(f'dataset/dataset/train/sphere/{i}')
    sizes_sphere.append(str(data.shape))
sizes_sphere = pd.Series(sizes_sphere)
print(sizes_sphere.value_counts())
```

```python
sizes_vort = []
for i in tqdm(os.listdir('dataset/dataset/train/vort')):
    data = np.load(f'dataset/dataset/train/sphere/{i}')
    sizes_vort.append(str(data.shape))
sizes_vort = pd.Series(sizes_vort)
print(sizes_vort.value_counts())
```

Therefore, all the images are of (150x150) dimension with a single channel.
