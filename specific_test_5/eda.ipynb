# Imports

```python
# Importing required modules
import warnings
import os
import seaborn as sns
from tqdm import tqdm
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

# EDA

### Class count

```python
counts = {}
print(os.listdir('dataset/lenses'))
for i in os.listdir('dataset/lenses'):
    counts[i] = len(os.listdir(f'dataset/lenses/{i}'))
```

```python
sns.barplot(x = list(counts.keys()), y = list(counts.values()))
```

The classes are perfectly balanced.

### Unique sizes

```python
sizes = []
for i in tqdm(os.listdir('dataset/lenses/sub')):
    data = cv2.imread(f'dataset/lenses/sub/{i}')
    sizes.append(str(data.shape))
sizes = pd.Series(sizes)
print(sizes.value_counts())
```

```python
sizes_sphere = []
for i in tqdm(os.listdir('dataset/lenses/no_sub')):
    data = cv2.imread(f'dataset/lenses/no_sub/{i}')
    sizes_sphere.append(str(data.shape))
sizes_sphere = pd.Series(sizes_sphere)
print(sizes_sphere.value_counts())
```

Therefore, all the images are of (150x150) dimension with a single channel.
